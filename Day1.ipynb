## 📢 Part 7: Capstone Project - Archery Game Setup 🎯

Let’s start building your Archery Game!

```python
import pygame
import sys

pygame.init()
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Archery Game 🎯")

try:
    background = pygame.image.load('sprites/sunset.png')
    archer = pygame.image.load('sprites/archer.png')
except:
    print("Error loading images! Check your folders.")

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.blit(background, (0, 0))
    screen.blit(archer, (WIDTH//2 - 50, HEIGHT - 120))
    pygame.display.flip()

pygame.quit()
```

🎉 **Congrats! Your archery game window now shows a background and archer sprite!**
